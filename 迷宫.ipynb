```python
import pandas as pd
import numpy as np
Map = pd.read_csv("./map.csv",index_col=0)
maze = np.array(Map)
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
count=0
if route_stack[-1] == [15,15]:
    count+=1
print(count)


maze[10,11]=1
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)


maze[10,7]=1
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)

maze[9,12]=1
maze[10,11]=0
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)






Map = pd.read_csv("./map.csv",index_col=0)
maze = np.array(Map)
route_stack = [[0,0]]
route_history = [[0,0]]
maze[6,2]=1
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)


maze[10,11]=1
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)


maze[10,7]=1
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)

maze[9,12]=1
maze[10,11]=0
route_stack = [[0,0]]
route_history = [[0,0]]
source=maze
def up(location):
    #横坐标为0，无法再向上走
    if location[1] == 0:
        return False
    else:
        new_location = [location[0],location[1]-1]
        #已经尝试过的点不会尝试第二次
        if new_location in route_history:
            return False
        #碰到墙不走
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def down(location):
    if location[1] == 15:
        return False
    else:
        new_location = [location[0],location[1]+1]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
 
def left(location):
    if location[0] == 0:
        return False
    else:
        new_location = [location[0]-1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
def right(location):
    if location[0] == 15:
        return False
    else:
        new_location = [location[0]+1,location[1]]
        if new_location in route_history:
            return False
        elif source[new_location[0]][new_location[1]] == 1:
            return False
        else:
            route_stack.append(new_location)
            route_history.append(new_location)
            return True
lo = [0,0]
while route_stack[-1] != [15,15]:
    if up(lo):
        lo = route_stack[-1]
        continue
    if left(lo):
        lo = route_stack[-1]
        continue
    if right(lo):
        lo = route_stack[-1]
        continue
    if down(lo):
        lo = route_stack[-1]
        continue
    route_stack.pop()
    lo = route_stack[-1]
if route_stack[-1] == [15,15]:
    count+=1
print(count)
```
