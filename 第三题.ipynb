```python
import numpy as np
import pandas as pd
import talib as ta
import matplotlib.pyplot as plt
data = pd.read_excel('sz50.xlsx', sheetname='600036.XSHG', index_col='datetime')
arr=data['close']
arr1 = np.array(arr)
s = pd.Series(ta.MA(arr1, 10))
new_index = pd.date_range("20170103",periods=len(s), freq="1.5D")
s.index = new_index
print(s)
plt.figure(figsize=(15, 7))
plt.plot(s)
plt.plot(data.close)
plt.show()
```
